Importa-se as bibliotecas pandas, numpy, seaborn, matplotlib.pyplot e chamou-se pd, np, sns e plt respetivamente.

```python
data = data.drop('Id',axis=1)
data
```

```python
data_semNA = data.dropna()
data_semNA
```

Atribui-se à variável x o *dataframe* excluindo a coluna 'SalePrice'.<br>
Atribui-se à variável y a coluna 'SalePrice'.

```python
x = data_semNA.iloc[:, :-1].values
y = data_semNA.iloc[:, 5].values
```

Importa-se o pacote train_test_split da biblioteca sklearn.model_selection.

```python
from sklearn.model_selection import train_test_split
```

Divide-se em *arrays* as variáveis x e y, onde o teste vai conter 20% do train, atribui-se um número a este teste.<br>
Mostra-se a forma dos *arrays*.

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

Importa-se o pacote LinearRegression da biblioteca sklearn.linear_model.

```python
from sklearn.linear_model import LinearRegression
```

Atribui-se à variável rlm o método LinearRegression().

```python
rlm = LinearRegression()
```

Ajusta-se o *array* ao modelo.

```python
rlm.fit(x_train,y_train)
```

Apresenta-se a interseção e os coeficientes.

```python
print('Intercept: \n', rlm.intercept_)
print('Coefficients: \n', rlm.coef_)
```

Apresenta-se a exatidão do test e do train.

```python
print("Accuracy test  :", rlm.score(x_test, y_test))
print("Accuracy train :", rlm.score(x_train, y_train))
```

Atribui-se à variável predictions a previsão do x_test.

```python
predictions = rlm.predict(x_test)
predictions
```

Compara-se o valor Atual com o valor previsto.

```python
A_Pred = pd.DataFrame({'Atual': y_test, 'Predicted': predictions})
A_Pred
```

Importa-se o pacote metrics da biblioteca sklearn.

```python
from sklearn import metrics
```

Apresenta-se os valores métricos com os valores previstos com o y_test.

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, predictions))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, predictions))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
